## Ejercicio 1: Funciones


De nuestro código que evaluaba cual era una el número mayor, convertirla a función.

Entrada: num1, num2

Salida: 

1. El num1 es mayor que num2
2. El num2 es mayor que num2
3. Ambos son iguales

````python
def suma(num1, num2) :

    if num1 > num2 :
        print(f"El {num1} es mayor que {num2}")

    elif num2 > num1:
        print(f"El {num2} es mayor que {num1}")
    else:
        print("Ambos son iguales")

num1= int(input("Dame el primer número"))
num2= int(input("Dame el segundo número"))

suma(num1,num2)
````

## Extra points

Hacer una función que nos imprima cuantos números pares e impares tiene una lista.

Para evaluar un par o impar puedes utilizar el

```Python
if numero%2 = 0:
    print("numero es par")
else:
    print("numero es impar")
```

|Entrada | Salida |
| - | - | 
| [2,3,6,8,9] | Tiene `3` pares y `2` inpares |

````python
lista = [2,3,6,8,9]

def pares_impares(lista):
    pares,impares = 0,0
    for num in lista:
        if num % 2 == 0:
            pares += 1
        else:
            impares += 1
    return pares, impares
            
x = pares_impares(lista)
print(f"Tiene {x[0]} pares y {x[1]} inpares")

````
